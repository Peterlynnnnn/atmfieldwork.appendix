```python
# importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## DATA ASSIMILATION: OPTIMAL INTERPOLATION

```python
#observation data( in degree celcius)
observation= np.array([15.2, 16.1, 14.5, 15.8, 25.0 ])

#model_output data( in degree celcius)
model_output= np.array([14.8, 15.5, 14.0, 16.0, 25.3 ])

    #PERFORMING OPTIMAL INTERPOLATION

# 1. Calculating the weights of both dataset

# calculation of variances

var_observation = np.var(observation)

var_modeling = np.var(model_output)

# Calculating for required weights

W_observation = 1 / var_observation / (1 / var_observation + 1 / var_modeling)

W_modeling = 1/ var_modeling / (1 / var_observation + 1 / var_modeling)

# 2. Calculation of Optimal Interpolation
analyzed_data = W_observation*observation + W_modeling*model_output

#3. Calculation of root mean squared error(rsme)
rmserror = np.sqrt(np.mean((analyzed_data-observation)**2))

#4. Calculation of bias error
bias = np.mean(analyzed_data-observation)

# Printing results
print("Model_output: ", model_output)
print("Observations: ", observation)
print("Weight_model: ", W_modeling)
print("Weight_observation: ", W_observation)
print("Analyzed_data: ", analyzed_data)
print("RMSERROR: ", rmserror)
print("Bias error: ", bias)
```

## PLOTTING COMPARISION DATASET

```python
#comparing analyzed data with Observation data
timesteps = ([1, 2, 3, 4, 5])
observation = ([15.2, 16.1, 14.5, 15.8, 25.0])
model_output = ([14.8, 15.5, 14.0, 16.0, 25.3])
Analyzed_data =([15.0, 15.8, 14.3, 15.9, 25.1])

# plotting of figure
fig = plt.subplots(figsize = (15,10))
plt.plot(timesteps,observation,label="observation", color= "green", marker = "o")
plt.plot(timesteps,model_output, label = "model", color = "red", marker =".")
plt.plot(timesteps,Analyzed_data, label = "Analysis", color = "black", marker = "^")
plt.legend(loc = 'upper center')
plt.grid(color = 'black')
plt.title("COMPARING ANALYSIS BETWEEN DATASETS", fontweight = "bold")
plt.xlabel("TIMESTEPS", fontweight = "bold")
plt.ylabel("TEMPERATURE(°C)", fontweight = "bold")
plt.show()
```
